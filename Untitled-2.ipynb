```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error
from sklearn import model_selection
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_spx = pd.read_csv('data/SPX.txt', sep='\t', header=None)
```

```python
#Data preparation
df_spx=df_spx.rename(columns={0:'Date', 1:'Cours'})
df_spx['Date']=pd.to_datetime(df_spx['Date'], format='%d/%m/%Y')
df_spx = df_spx.set_index('Date')
df_spx = df_spx.sort_index()
df_spx['Cours']=pd.to_numeric(df_spx['Cours'].str.replace(',','.'))

#valeurs nulles
print(' valeurs nulles:',df_spx.isnull().sum().sum())
df_spx.head()
```

```python
# diagnostic de la série temporelle

plt.plot(df_spx['Date'], df_spx['Cours'], label ='Cours SPX') 
plt.ylabel('Cours')
plt.xlabel('Date')

plt.legend('evolution du cours du SPX')
plt.show()
```

```python
#Séparation des données en données d'entrainement et de test
train = df_spx[df_spx.index < pd.to_datetime("2012-01-01", format='%Y-%m-%d')]
test = df_spx[df_spx.index >= pd.to_datetime("2012-01-01", format='%Y-%m-%d')]
```

```python
plt.plot(train, color = "black")
plt.plot(test, color = "red")
plt.ylabel('Cours spx')
plt.xlabel('Date')
plt.xticks(rotation=45)
plt.title("Train/Test pour spx")
plt.show()
```
